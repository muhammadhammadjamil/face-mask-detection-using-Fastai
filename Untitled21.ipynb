```python
path=path/'/content/drive/MyDrive/face/dest_folder'
path.ls()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data=ImageDataBunch.from_folder(path,train='train', valid='val', ds_tfms=get_transforms(),size=224)
```

```python
data.normalize(imagenet_stats)
```

```python
data.show_batch()
```

```python
learn=cnn_learner(data, models.resnet50,metrics=[accuracy,error_rate])
```

```python
learn.summary()
```

```python
learn.fit_one_cycle(4)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix()
```

```python
i=open_image('/content/drive/MyDrive/face/dest_folder/test/with_mask/114-with-mask.jpg')
i
```

```python
what,_,probs =learn.predict(i)
print(f"It's a {what}!")
print(f"Probability it's a {what}: {probs[1].item():.6f}")
```
